# arrays

Un array es un area contigua de memory separada en elementos con el mismo tamano. Y cada elemento tiene un indice 0,1,2,3,4,.... 


Los Arrays tienen Constant time acces. Es decir, acceder a cualquier elemento guardado en un array por medio del indice se puede lograr sin perder tiempo. Ya que con s'olo conocer el 'indice, como todos los elementos del array tienen el mismo tamano, se puede calcular la posicion en la memoria del elemento con array con 'indice que estamos buscando.

Las operaciones de agregar o quitar elemento al final son muy rapidas en un array, pero agregar o quitar en medio es muy lento porque hay que reordenar todo


# linked List



En este caso, cada elemento del list contiene un key (el valor) pero tambien contiene un pointer al lugar de memoria en la que se encuentra el siguiente elemento. 


Los tiempos de estas operaciones pueden ser largos o cortos. Hay que tomar en cuenta que en cada operacion, hay que actualizar los pointers, por lo que puede tomar mucho tiempo. Hay que tomar en cuenta que para llegar a el i-esimo elemento de la lista, no tenemos constant time access, sino que hay que ir recorriendo la lista desde el principio usando los pointers hasta llegar al i-esimo. Se puede agregar un tail pointer que apunta directo al ultimo elemento. 


#  Doubly Linked List

En un doubly linked list, cada elemento tiene dos pointers, uno a la posicion en memoria del siguiente elemento y uno a la posicion del anterior. Esto puede hacer ciertas operaciones mas rapidas. 


# Stacks

Es un tipo de data con tres operaciones: <br> 1) Push(key) que agrega un elemento al final. <br> 2) Top() Regresa el valor del ultimo elemento <br> 3) Pop(): Regresa el valor del ultimo  elemento y lo quita <br>
Se puede pensar como una columna de libros por ejemplo

Ejemplo: Balanced Brackets

```python
#Buscamos comprobar si una lista de de parentesis y corchetes est'a balanceada 
#(todo abierto se cierra)

def IsBalanced(stri):
    #voy a usar un tipico array como stack, con pop y append.
    stack=[]
    
    #vamos agarrando cada elemento del str
    for char in stri:
        #Si es un apridor, lo agregamos al stack de abridores y ya
        if char in ['[','(']:
            stack.append(char)
        
        # Si no es un abridor (es un cerrador)
        else:
            #Si no hay elementos en el stack de abridores
            if len(stack) ==0:
                #significa que el cerrador era inbalanceado, por lo que false
                return(False)
            
            # si s'i hay un elemento en el stack de abridores, le llamamos 
            # top y lo sacamos
            top = stack.pop()
            
            
            # Si el ultimo cerrador (que es char) no coincide con el ultimo
            # abridor que no ha sido cerrado (que es top), entonces esta mal
            if (top == '[' and char != ']') or (top=='(' and char != ')'):
                return(False)
    # Una vez recorrda toda la lista, si quedan elementos en el stack
    # significa que no todos fueron cerrados y hay que regresar false. 
    # La unica forma de regresar true es que el stack este vacio, mostrando
    # que todos los abridores fueron cerrados
    return (len(stack) ==0)
            
print(IsBalanced('()[](())'))
            
            
```

# Queue

Es un abstrac data type con las siguientes operaciones: <br> 
1) enqueue(key): Agrega la key al final de la coleccion <br>
2) Dequeue(): <br> quita el elemento mas viejo y lo devuelve <br>
Un queue saca el elemento mas viejo, el que lleva mas esperando. Es como una fila de personas. 

Implementarla: Se puede implementar con una lista con con doble linked. Para enqueue, agrega algo y le pone el pointer de tail. Para deque, mueve el pointer de head al siguente elemento.<br>
Implementarlo en un array no es tan buena idea, ya que quitar el primer elemento seria muy costoso, aunque se puede ir guardando la cantidad de elemento para usar el array como queue

```python
#En python
# se llama deque y tiene append y popleft
from collections import deque
queue = deque(["Eric", "John", "Michael"])
queue.append("Terry")           # Terry arrives
queue.append("Graham")          # Graham arrives
queue.popleft()                 # The first to arrive now leaves

queue.popleft()  # The second to arrive now leaves
print(queue)
```

Level(ana): Es la fila en la que se encuentra la key (ana) <br>
Height(Cathy): Es el numero de niveles mas abajo en el que se enceuntra el ultimo descendiente de Cathy 
##### Binary Search Tree: 
Cada nodo  tiene a lo sumo dos hijos. Y a parte, para cada nodo, los elementos a la izquierda son menores y a la derecha son mayores. Permite hacer un binary search muy rapido
 <br> 


Un binary tree solo tiene hijos izquierdo y derecho en cada nodo. <br>
Se puede usar un array con pointers a los hijos (y si se quiere a los padres) para representar un tree

### Transversar un Tree:

Hay dos formas de hacerlo: <br> 
###### 1) Depth First:
Viajamos por completo en un subtree antes de pasar al hermano del nodo <br> 
1.1) InOrder: Para cada nodo, vamos exploramos el hijo izq, al terminar de explorar imprimios el nodo y luego exploramos el derecho. 

En un tree binario, este tipo de busqueda puede servir para imprimir los elementos en orden. <br>

1.2) PreOrder: Empezamos en el root y lo imprimimos, luego aplicamos el preorder al hijo izquierdo y al terminar, al hijo derecho (o a todos los hijos si hay mas). <br> 
Entonces, va a imprimir el root y luego los izquierdos y cada que termina de investigar todos los izquierdos de un nodo, se va al lado derehco. 


1.3) Pos Order: Primero investigamos el izquierdo, luego el derecho y al final imprimimos. El Root sera el ultimo en imprimir.


###### 2) Breadth First: 
Transversamos todos los nodos de un nivel antes de pasar al siguiente


###### 1. Checar brackets en un codigo

El input trae muchos () [] y tenemos que decir si todo abierto se cierra o si hay algun problema. 

```python
#Checa si hay algun error en una expresion con parentesis y breackets.
''' Si todo est'a bien, imprime success
Si hay un error, si sobra un cerrado, imprime su indice
si falta un cerrado, imprime el primer abierto que no tiene cerrador'''
#Usaremos un stack porque vamos a hacer (Last in first out)


open_list = ["[","{","("]
close_list = ["]","}",")"]

# Toma un bracket cerrador y lo hace abridor
def cambia(k):
    if k == "]":
        return "["
    elif k == ")":
        return "("
    elif k == "}":
        return "{"


def checa(mystr):
    global stack
    #Stack guarda todos los abridores
    #stack num guarda los indices de los abridores guardados
    stack=[]
    stacknum=[]


    #Itera sobre los caracteres uno a uno.
    for i in range (len(mystr)):
        #Guarda el caracter i-esimo en curr
        curr = mystr[i]
        #Si dicho caracter es un abridor
        if curr in open_list:
            #Guarda el abridor al final de stack y tambien guarda su numero por si las dudas
            stack.append(curr)
            stacknum.append(i)

        #Si el caracter es cerrador, entonces
        elif curr in close_list:
            
            #Si ya no queda nada en stack, y aun asi hay un cerrador,
            #Significa que dicho cerrador est'a de m'as
            #Por lo tanto, imprime su indice (m'as uno porque as'i lo pide el ej.)
            if len(stack) == 0:
                return(i+1)

            # Si el ultimo elemento en stack (el ultimo abridor) coincide
            #con el cerrador que acabamos de encontrar, significa que se cancelan
            # Y entonces podemos sacarlo del stack
            #y tambien sacamos su numero.
            if stack[-1] == cambia(curr):
                stack.pop(-1)
                stacknum.pop(-1)
            else:
                return(i+1)
            
    #Si no queda nada en el stack es porque todos los brackets fueron cerrados
    if len(stack) == 0:
        return "Success"
    #Si queda algo, es porque hay un abierto que no se cerr'o,
    #El primer abierto que no se cerro esta guardado en stacknum[0]
    else:
        return(stacknum[0]+1)
    

text = list(input())
print(checa(text))

            
```

```python
''' Nos dan una serie de inputs, si dicen push key, hay que agregar la key a una lista
Si dice pop, hay que poppear el ultimo y si dice max hay que imprimir el max
de la lista hasta el momento'''

import sys


class StackWithMax():
    #Un elemento en esta clase tiene dos stacks (el normal y el aux)
    def __init__(self):
        self.__stack = []
        self.__aux_stack = []

    def push(self, a):
        #Agrega el elemento al stack
        self.__stack.append(a)
        #Si el auxiliar es vacio, le agrega a tambien
        if len(self.__aux_stack) == 0:
            self.__aux_stack.append(a)
        #Si a es mas grande que el ultimo elemento del aux, le agrega a tambien
        elif a >= self.peek_aux():
            self.__aux_stack.append(a)
        #Es decir, solo agrega la a al aux si es mas grande que la ultima o si esta vacio

    def peek_aux(self):
        #Peek aux checa si el aux es no vacio
        #Y regresa el ultimo elemento del aux
        assert(len(self.__stack))
        return self.__aux_stack[-1]

    def pop(self):
        #Pop checa si el stack tiene elementos
        assert(len(self.__stack))
        #Y checa si el ultimo elemento del aux es igual al ultimo del stack
        #Si lo es, elimina el ultimo elemento del aux
        #Porque significa que hemos perdido ese maximo
        if self.__stack.pop() == self.peek_aux():
            self.__aux_stack.pop()
        
    def max(self):
        #Checa si el stack es no vacio
        assert(len(self.__stack))
        #Aplica la funcion peek auz
        return self.peek_aux()


if __name__ == '__main__':
    #Creamos un elemento de la clase
    stack = StackWithMax()
    
    #Medio extrana forma de tomar el input()
    num_queries = int(sys.stdin.readline())
    
    for _ in range(num_queries):
        #query toma el input actual
        query = sys.stdin.readline().split()

        #Si la primera palabra del input es push, agregamos la segunda palabra a la lista usando la funcion creada push
        if query[0] == "push":
            stack.push(int(query[1]))
        elif query[0] == "pop":
        #Si la entrada dice pop, sacamos el ultimo elemento usando la funcion de la clas           
            stack.pop()
        elif query[0] == "max":
        #Si dice max, aplicamos la funcion max de la clase
            print(stack.max())
        else:
            assert 0


''' Basicamente, al stack le realizamos todas las operaciones,
pero al aux solo le agregamos los elementos que se puedan agregar en orden ascendente,
para tener siemore el maximo del stack al final del aux
Luego, si hacemos un pop, nos fijamos si este pop elimino del stack al ultimo del aux
en cuyo caso lo quitamos del aux porque ya no es el mas grande '''
```

###### 3. Processing

```python
''' Network packet processing simulation '''
#LLega informacion sobre unos paquetes que hay que procesar.
#La computdadora tiene un buffer donde puede guardar los paquetes antes de procesarloes
#La primera linea del input es el tamano del buffer y el numero de elementos.
#A partir de ahilas siguientes lineas son los tiempos de llegada de los trebajos
# Y sus tiempos de procesamiento necesario.
#Outputear en lineas el momento en el que el procesador empezo a trabajar o un -1 si no lo proceso



class Request:
    def __init__(self, arrival_time, process_time):
        self.arrival_time = arrival_time
        self.process_time = process_time

class Response:
    def __init__(self, dropped, start_time):
        self.dropped = dropped
        self.start_time = start_time

class Buffer:
    def __init__(self, size):
        self.size = size
        self.finish_time = []

    def Process(self, request):
        # remove all finished job (has finish_time < current_request.start_time
        while len(self.finish_time) > 0 and self.finish_time[0] <= request.arrival_time :
            self.finish_time.pop(0)

        if len(self.finish_time) < self.size:
            if len(self.finish_time) == 0:
                self.finish_time.append(request.arrival_time + request.process_time)
                return Response(True, request.arrival_time)
            else:
                start_time = request.arrival_time
                if self.finish_time[-1] > start_time:
                    start_time = self.finish_time[-1]
                elif self.finish_time[-1] == start_time:
                    start_time = self.finish_time[-1] + 1
                self.finish_time.append(start_time + request.process_time)
                return Response(True, start_time)
        else:
            return Response(False, -1)

def ReadRequests(count):
    requests = []
    for i in range(count):
        arrival_time, process_time = map(int, input().strip().split())
        requests.append(Request(arrival_time, process_time))
    return requests

def ProcessRequests(requests, buffer):
    responses = []
    for request in requests:
        responses.append(buffer.Process(request))
    return responses

def PrintResponses(responses):
    for response in responses:
        print(response.start_time if response.dropped else -1)

if __name__ == "__main__":
    size, count = map(int, input().strip().split())
    requests = ReadRequests(count)

    buffer = Buffer(size)
    responses = ProcessRequests(requests, buffer)

    PrintResponses(responses)
```

###### Max window

```python
from collections import deque 
  
# A Deque (Double ended queue) based  
# method for printing maximum element  
# of all subarrays of size k  
def printMax(arr, n, k): 
      
    """ Create a Double Ended Queue, Qi that  
    will store indexes of array elements.  
    The queue will store indexes of useful  
    elements in every window and it will 
    maintain decreasing order of values from 
    front to rear in Qi, i.e., arr[Qi.front[]] 
    to arr[Qi.rear()] are sorted in decreasing 
    order"""
    Qi = deque() 
      
    # Process first k (or first window)  
    # elements of array 
    for i in range(k): 
        
        # For every element, the previous  
        # smaller elements are useless 
        # so remove them from Qi 
        while Qi and arr[i] >= arr[Qi[-1]] : 
            Qi.pop() 
          
        # Add new element at rear of queue 
        Qi.append(i); 
          
    # Process rest of the elements, i.e.  
    # from arr[k] to arr[n-1] 
    for i in range(k, n): 
          
        # The element at the front of the 
        # queue is the largest element of 
        # previous window, so print it 
        print(str(arr[Qi[0]]) + " ", end = "") 
          
        # Remove the elements which are  
        # out of this window 
        while Qi and Qi[0] <= i-k: 
              
            # remove from front of deque 
            Qi.popleft()  
          
        # Remove all elements smaller than 
        # the currently being added element  
        # (Remove useless elements) 
        while Qi and arr[i] >= arr[Qi[-1]] : 
            Qi.pop() 
          
        # Add current element at the rear of Qi 
        Qi.append(i) 
      
    # Print the maximum element of last window 
    print(str(arr[Qi[0]])) 
      
# Driver programm to test above fumctions 
if __name__=="__main__":
    n = int(input())
    arr = [int(x) for x in input().split()]
    m=int(input())
    
    printMax(arr, n, m) 
      


    
```

##### Tree Height

```python
import sys
import threading

sys.setrecursionlimit(10 ** 7)  # max depth of recursion
threading.stack_size(2 ** 25)  # new thread will get stack of such size


class TreeHeight:

    def __init__(self):
        self.n = 0
        self.parent = []
        self.cache = []

    def read(self):
        """Reads data from standard input."""
        self.n = int(sys.stdin.readline())
        self.parent = list(map(int, sys.stdin.readline().split()))
        self.cache = [0] * self.n

    def path_len(self, node_id):
        """Returns path length from given node to the root."""
        parent = self.parent[node_id]
        if parent == -1:
            return 1
        #Si existe algun elemento en el cache, lo regresa
        if self.cache[node_id]:
            return self.cache[node_id]

        #Si no existe, lo agrega
        #hace 1 + el path length del pap'a 
        self.cache[node_id] = 1 + self.path_len(self.parent[node_id])
        return self.cache[node_id]

    def compute_height(self):
        """Computes the tree height."""
        return max([self.path_len(i) for i in range(self.n)])


def main():
    tree = TreeHeight()
    tree.read()
    print(tree.compute_height())


threading.Thread(target=main).start()
```
