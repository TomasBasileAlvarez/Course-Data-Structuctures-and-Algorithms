## Binary Search Trees

Buscamos una data structure que permita encontrar elementos de una b'usqueda o al menos encontrar elementos cercanos a lo que buscamos. <br>
Adem'as tiene que ser din'amica, permitiendo agregar y quitar elementos. <br> <br> 

Intentos: <br> 
1) Array: No es buena idea porque toma un tiempo O(n) hacer las cosas. <br>
2) Sorted Array: Buscar cosas toma un tiempo O(log(n)) pero insertar toma un tiempo O(n) porque hay que recorrer los elementos del array, lo mismo con delete. <br> 
3) Linked List: Toma un tiempo O(n) encontrar elementos. <br><br>
Necesitamos encontrar algo mejor

#### Search Trees

Un search tree es un arbol con a lo sumo dos hijos en cada nodo en el que el izquierdo es menor y el derecho es mayor. <br> 
Y de hecho, todos los elementos a la derecha de un nodo son mayores y todos a la izquierda son menores. <br> 
Cada elemento del binary tree tiene propiedades R.key (su valor), R.left, R.right los hijos izquierdo y derecho

###### Operaciones Basicas


Empezamos con un arbol como el de arriba y vamos recorriendo el arbol segun corresponda.



###### Find(k)
Dado un arbol y un valor k que buscamos, podemos buscarlo como sigue

Si k esta en el arbol, el find regresa dicho valor k, si no est'a, en este caso no regresa nada. Sin embargo, haber recorrido el arbol nos dej'o en el lugar en el que ir'ia k. 

##### Next(N) 
Dado un nodo N, buscamos el nodo con el siguiente elemento m'as grande. Claramente, tenemos que ir a la derecha y luego a la izquierda muchas veces. Si no se puede ir a la derecha, tenemos que ir subiendo lo m'as que se pueda


###### Descendiente izquierdo: 
Busca el descendiente izquiedo m'as abajo


###### Right Ancestor:
Va buscando el padre derecho m'as arriba


#### Range Search

Dados dos numeros x,y y un Root R, buscamos todos los nodos entre x y y.


##### Insert(k)
Basicamenta hay que ir recorriendo el arbol con un find y donde nos detengamos es donde debe de ir el nodo con key k que queremos meter.


##### Delete (N)
