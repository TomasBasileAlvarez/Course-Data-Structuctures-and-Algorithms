# Algoritmo A estrella
Es una mejora al algoritmo de Dijsktra doble que implementa una aproximacion inicial de las distancias. <br> 

##### Potential Function
Una funcion potencial pi(v) mapea los vertices a numeros reales. Luego, si reemplazamos la distancia $l(u,v)$ entre los vertices u,v por la funcion $l_{\pi}(u,v) = l(u,v) + \pi(v) - \pi(u)$.<br>
Un lema nos dice que el mejor camino demostrado por Dijsktra es igual si cambiamos l por l_pi. <br> <br>

$\textbf{Lema:} $ Para una funcion potencial $pi: V \rightarrow R $, para cualesquiera dos vertices s,t y un camino P se cumple que la longitud del camino es $l_{pi}(P) = l(P) - \pi(s) + \pi(t)$. 


Tambien se puede hacer un Astar bidireccional como con Dijsktra
